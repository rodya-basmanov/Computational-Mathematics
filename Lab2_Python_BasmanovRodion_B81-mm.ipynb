```python
x = sp.symbols('x')

expr = x**3 - 9 * x**2 + 23 * x - 15

eq = sp.Eq(expr, 0)

left = eq.lhs

right = eq.rhs

solved_eq = sp.solve(eq, x)

print(solved_eq)

value = [left.subs({x : i}) for i in solved_eq]

print(value)

graf_12 = sp.plot(left, right, xlim = (0, 6), ylim = (-5, 5), title = "Chart 1 and 2", xlabel = "x", ylabel = "y", legend = True, show = False)

graf_12[0].line_color = "blue"
graf_12[0].label = "y = x**3 - 9 * x**2 + 23 * x - 15"

graf_12[1].line_color = "red"
graf_12[1].label = "y = 0"

graf_12.show()
```

# 8

```python
x = sp.symbols('x')

expr = x**3 - 9 * x**2 + 23 * x - 15

eq = sp.Eq(expr, 0)

left = eq.lhs

f_diff = sp.diff(left, x)

a, b = 1, 5

critical_points = sp.solve(f_diff, x)

critical_points_in_interval = [cp for cp in critical_points if a <= cp <= b]

candidates = critical_points_in_interval + [a, b]

values = [(point, float(left.subs({x : point}).evalf())) for point in candidates]

min_point, min_value = min(values, key=lambda item: item[1])

max_point, max_value = max(values, key=lambda item: item[1])

min_point.evalf()
```
